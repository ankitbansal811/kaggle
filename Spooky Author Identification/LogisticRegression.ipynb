```python
data_folder = Path('data')
train_zip = 'train.zip'
test_zip = 'test.zip'
```

```python
train = read_from_zip(data_folder.joinpath(train_zip))
test = read_from_zip(data_folder.joinpath(test_zip))
```

```python
train.head()
```

```python
X = train['text']
y = train['author']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
import xgboost
```

```python
pipeline = Pipeline([
    ('feature', TfidfVectorizer()),
    ('clf', LogisticRegression(max_iter=200)),
])

params_grid = {
    'feature__ngram_range' : [(1,1), (1,2), (1,3)],
    'feature__min_df' : [1, 3, 5],
#     'feature__stop_words' : ['english', None],
#     'feature__max_features' : [None, 5000, 10000, 25000],
    'clf__C' : [0.1, 0.5, 1, 1.5],
}
```

```python
grid_clf = GridSearchCV(estimator=pipeline, param_grid=params_grid, verbose=1000, scoring='neg_log_loss', n_jobs=7, cv=3)
```

```python
grid_clf.fit(X, y)
```

```python
len(grid_clf.best_estimator_['feature'].vocabulary_)
```
